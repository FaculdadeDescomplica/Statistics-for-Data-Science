# `Probabilidade`

#### A definio clssica de probabilidade 




Ou seja, probabilidade seria a diviso do nmero de casos favorveis (definido pelo evento A) pelo nmero de casos totais (definido pelo espao amostral).




### Uma probabilidade tem vrias propriedades. Vejamos as mais importantes!


<br>
<br>
<br>

* A probabilidade de um evento A acontecer est sempre entre 0 e 1


* A probabilidade do espao amostral $\Omega$ ocorrer  1


* A probabilidade de um evento $\emptyset$, que esteja fora do espao amostral, ocorrer  0


* Se $\bar{A}$ for o evento complementar de $A$, ento $P(\bar{A}) = 1 - P(A)$

Dado dois eventos A e B, a probabilidade da unio de A e B ser

### $P(A \cup B) = P(A) + P(B) - P(A \cap B)$

#### Mutuamente Exclusivos

Dizemos que dois eventos so mutuamente exclusivos quando a probabilidade da interseo entre eles  igual a zero.

![alt_text](https://s4.static.brasilescola.uol.com.br/be/2020/06/4.jpg)

```python
# exemplo


p_ganhar_bonus_empresa = 0.4
p_ganhar_aumento = 0.25

p_ganhar_bonus_e_aumento = 0.1


p_ganhar_bonus_ou_aumento = p_ganhar_bonus_empresa + \
p_ganhar_aumento - p_ganhar_bonus_e_aumento

p_ganhar_bonus_ou_aumento
```

#### Eventos complementares

```python
p_chover_hoje = 0.15


p_nao_chover_hoje = 1 - p_chover_hoje
p_nao_chover_hoje
```
